# Kernel Regression Implementation 
Nadaraya–Watson kernel regression

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

# Data generating function
Used for generating data with or without noise

```python
def function(x):
     return np.sin(x)
```

```python
def find_y(x, is_noisy:bool=True, mean=0.0, sigma=0.75):
    if is_noisy:
        return function(x) + np.random.normal(mean,sigma)
    else:
        return function(x)
```

```python
def generate_data(left_bound, right_bound, number_of_points=100, is_noisy:bool=True):
    points_x = np.linspace(left_bound, right_bound, num=number_of_points)
    points_y = [find_y(x, is_noisy) for x in points_x]
    return points_x, points_y
```

#### Uniform Kernel $[|r| \leq 1]$

```python
def uniform_kernel(x, x0, h=1.0) -> np.ndarray:
    # initialize weights
    weights = np.zeros(x.shape)
    r = np.abs((x - x0)/h)
    
    # find where r <= 1
    indx = np.where(r <= 1)
    weights[indx] = 1
    return weights
    
```

#### Triangle Kernel $(1 - |r|)\;[|r|\leq1]$

```python
def triangle_kernel(x, x0, h=1.0) -> np.ndarray:
    # initialize weights
    weights = np.zeros(x.shape)
    r = np.abs((x - x0)/h)
    
    # find where r <= 1
    indx = np.where(r <= 1)
   
    weights[indx] = (1 - r[indx])
    return weights
```

#### Epanechnikov Kernel $(1- r^2)\;[|r|\leq1]$

```python
def epanechnikov_kernel(x, x0, h=1.0) -> np.ndarray:
    # initialize weights
    weights = np.zeros(x.shape)
    r_abs = np.abs((x - x0)/h)
    r  = (x - x0) / h
    
    # find where r <= 1
    indx = np.where(r_abs <= 1)
    weights[indx] = (1 - r[indx]**2)
    return weights
```

#### Quartic Kernel $(1-r^2)^2\;[|r|\leq1]$

```python
def quartic_kernel(x, x0, h=1.0) -> np.ndarray:
    # initialize weights
    weights = np.zeros(x.shape)
    r = np.abs((x - x0)/h)
    
    # find where r <= 1
    indx = np.where(r <= 1)
    weights[indx] = (1 - r[indx]**2)**2
    return weights
```

#### Gaussian Kernel $exp(-2r^2)$

```python
def gaussian_kernel(x, x0, h=1.0) -> np.ndarray:
    # initialize weights
    weights = np.zeros(x.shape)
    r = (x - x0) / h
    
    weights = np.exp(-2*r**2)
    
    return weights
```

## Nadaraya–Watson kernel regression

```python
def Nadaraya_Watson_regression(x, y, kernel=epanechnikov_kernel, h=0.5):
    length = len(x)
    y_est = np.zeros(length)
    
    # Initialize all weight from the choosen kernel function
    w = np.array([kernel(x, x[i], h) for i in range(length)]) 
    print(w.shape)
    
    # Nadaraya Watson 
    for i in range(length):
        y_est[i] = np.sum(w[i]*y) / np.sum(w[i])
        
    return y_est
```

## Data visualization

```python
x_noise, y_noise = generate_data(-10, 10, 1000)
x, y = generate_data(-10, 10, 1000, is_noisy=False)
y_est = Nadaraya_Watson_regression(x_noise, y_noise, kernel=epanechnikov_kernel, h=1)
```

```python
fig = plt.figure(figsize=[20,10])
plt.scatter(x_noise, y_noise, s=3)
plt.plot(x, y_est, c='red', label='N-W regression')
plt.plot(x, y, c='black', label='"True" regression')
plt.legend(loc="lower right")
```
